## 最終課題

```python
#武蔵野大学のWebサイトのサイトマップを抽出する
#武蔵野大学Webサイトのトップページにアクセス
#同一ドメインの全てのリンク（コメントアウトされていないもの）を辿り，全ページのURLと<title>を辞書型変数に格納する
```

```python
# webスクレイピングに最低限必要なライブラリをインポート
import requests
from bs4 import BeautifulSoup
import time
```

```python
# for 文等で特定のWebサイトに繰り返しアクセスする場合は、サーバーに負荷をかけないように適宜待機時間を入れる
for i in range(5):
    time.sleep(0.3)  #プログラムを0.3秒間停止
    print(f"アクセス回数：{i+1}")

#　アクセスしたい武蔵野大学のURLを指定
base_url = "https://www.musashino-u.ac.jp/"

#アクセスしたリンクを格納するリストを用意
accessed_link = []
#サイトマップ(URLとタイトル)を格納する辞書型変数を用意
sitemap = {}  




def crawl(url): #リンクを辿る関数を定義

     # 既にアクセス済みのものは飛ばす
    if url in accessed_link:
        return
    print(f"アクセスのURL: {url}") #アクセスしたURLを表示

    # アクセス済みとしてマーク
    accessed_link.append(url)

    try:
       #webサーバーにHTTPリクエストを送信
        #レスポンスを変数に格納しておく
        res = requests.get(url)
        res.encoding = res.apparent_encoding
        print(f"ステータスコード：{res.status_code}")  #ステータスコードを表示

        # HTMLのソースコードを解析
        soup = BeautifulSoup(res.text, 'html.parser')

        # ページのタイトルを取得してサイトマップに追加
        title = soup.title.string.strip() if soup.title else 'No title'
        sitemap[url] = title #URLとタイトルを辞書型変数に格納
        print(f'タイトル: {title}') #タイトルを表示

        #全てのaタグを抽出
        a_tag = soup.find_all('a')
        for a_tag in a_tag: #aタグを一つずつ処理
            print(a_tag.get('href'))  #href属性の値を表示 
            href = a_tag.get('href') #href属性の値を変数に格納
            if href:
                if href.startswith('https://www.musashino-u.ac.jp/') or href.startswith('/'):
                # 相対リンクを絶対リンクに変換
                    if href.startswith('/'):
                        full_href = base_url.rstrip('/') + href #絶対URLに変換
                    elif href.startswith('https://www.musashino-u.ac.jp/'):
                        full_href = href #絶対URLのまま
        
                # アクセスしていないリンクの場合
                    if full_href not in accessed_link:
                        time.sleep(0.1)  # サーバーに負荷をかけないように
                        
                        #再帰的にクロール
                        crawl(full_href)


    except Exception as e:
        print(f"エラーが発生しました:{url} {e}")            
#クロールする
crawl(base_url)
#集めたURLとタイトルを表示
print('Links and titles:')
for idx, (link, title) in enumerate(sitemap.items(), start=1): #番号付きで表示
    print(f'{idx}. {link}: {title}')  #URLとタイトルを表示

#一様ファイルに保存
with open('sitemap.txt', 'w', encoding='utf-8') as f:
    for url, title in sitemap.items():
        f.write(f"URL: {url}, Title: {title}\n")
```
